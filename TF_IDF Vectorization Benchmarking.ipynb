```python
import sys  
import pandas as pd
import numpy as np
sys.path.insert(0, '/Stats-Toolbox')

from statsbox import Decision_Tree
from statsbox import K_Means
from statsbox import TF_tools
df = pd.read_csv("Amazon-data.csv")
import nltk
nltk.download('stopwords')
```

### Preprocess text data  

```python
sentences = df['reviews.text']
x_sent = []
for s in sentences:
    x_sent.append(s)
x_sent=x_sent[500:1000]
```

```python
x = x_sent[20:120]
x = (TF_tools.remove_stopwords(TF_tools.lowercase(TF_tools.remove_punctuation(x))))
```

## Create corpus to input to our algorithm (Must be a list of strings) 

```python
x
```

#### use the helper function to get set of words

```python
words_set = TF_tools.get_word_set(x, 'set')
print(words_set)
```

## Generate the TF-IDF matrix for this corpus using our toolbox functions

```python
df_tf = TF_tools.TF_matrix(x, words_set, 'list')
df_tf_idf = TF_tools.TF_IDF(x, words_set, df_tf, 'list' )
```

```python
df_tf_idf = pd.DataFrame(df_tf_idf)
df_tf_idf
```

### Using sklearn.feature_extraction.TFidVectorizor on same data

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tr_idf_model  = TfidfVectorizer()
tf_idf_vector = tr_idf_model.fit_transform(x)
tf_idf_array = tf_idf_vector.toarray()
words_set = tr_idf_model.get_feature_names()

df_tf_idf_bench = pd.DataFrame(tf_idf_array, columns = words_set)
df_tf_idf_bench
```

## Benchmarking - compare the two tf-idf matrices 

### Using cosine similarity to rank the vectors 

### printing the cosine similarities of adjacent sentences in the corpus according to both matrices
##### we can see that the relative rank of the similarities is the same

```python
a=[]
b=[]
for i in range(8):
    a=(TF_tools.cos_sim(df_tf_idf_bench.iloc[i],df_tf_idf_bench.iloc[i+1] ))
    b=TF_tools.cos_sim(df_tf_idf.iloc[i],df_tf_idf.iloc[i+1] )

    print('cos similarity according to sklearn matrix ',a)
    print('cos similarity according to statsbox matrix ',b)
    print("")
```

### Using cosine similarity to find sentences similar to query sentence

#### create a query sentence

```python
query = "amazon fire is a great tablet to go online"
```

```python
x_set = x
x_set.append(query)
```

```python
words_set = TF_tools.get_word_set(x_set, 'set')
df_tf = TF_tools.TF_matrix(x_set, words_set, 'list')
df_tf_idf = TF_tools.TF_IDF(x_set, words_set, df_tf, 'list' )
```

```python
tr_idf_model  = TfidfVectorizer()
tf_idf_vector = tr_idf_model.fit_transform(x_set)
tf_idf_array = tf_idf_vector.toarray()
words_set = tr_idf_model.get_feature_names()

df_tf_idf_bench = pd.DataFrame(tf_idf_array, columns = words_set)
df_tf_idf_bench
```

```python
df_tf_idf = pd.DataFrame(df_tf_idf)
df_tf_idf
```

```python
a=[]
b=[]
for i in range(5):
    a.append(TF_tools.cos_sim(df_tf_idf_bench.iloc[i],df_tf_idf_bench.iloc[100] ))
    b.append(TF_tools.cos_sim(df_tf_idf.iloc[i],df_tf_idf.iloc[100] ))

print("index of most similar sentence acc. to sklearn is ",np.argmax(a))
print("index of most similar sentence acc. to statsbox is ",np.argmax(b))
```
